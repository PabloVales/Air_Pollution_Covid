# Polynomial linear regression working principle in sklearn

## How to do it?
Take PolynomialFeatures from sklearn and apply it to my data. If it is univariate (I only have 1 feature) then it will create $x$, $x^2$, $x^3$, etc. If I have 2 or more features it also adds all cross terms between them. This enhanced matrix X is then fed into the LinearRegression module from sklearn and that makes the polynomial fitting. 

## Why do we need to apply PolynomialFeatures before LinearRegression?
Ordinary least Squares (OLS, or MLE or whatever minimization function), cannot, by itself, find the correct equation with just feature X. If we look at the figure below for *degree 1* we see that it gives us a line that indeed minimizes the distance between the points. This is because OLS is a linear minimization function: we need to provide data to the desired degree to the model so that it can give the correct answer. 

## Key concepts:
LinearRegression uses each of the features as a separate dimension and executes a regression on it (to find, in conjuction, the variable $y$). The key idea is that LinearRegression does **always** a simple linear regression, because the **OLS** or whatever minimization function is, in itself, linear. Taking our particular example, if we apply PolynomialFeatures(degree=2) to our vector $x$, we will have a m x 2 matrix with $x$ and $x^2$. **Each** of these columns will be used to execute a simple linear regression. The former will give us almost 0, but the latter will have a **perfect** linear correlation, whose slope will be 1 in our case. In short: LinearRegression does simple linear regressions in **all** dimensions and the exponent component is embedded into the feature space. It's like making a "cambio de variable" in the equation from $y=x^2$ to $y=t$ (where $t=x^2$)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score


def true_fun(X):
    return 100*X**2

np.random.seed(0)

n_samples = 5
degrees = [1, 2, 4, 8, 50]

X = np.sort(np.linspace(-50, 50, num=n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

X, y
```

```python
plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(X[:, np.newaxis], y)
    
    print('Calculated features by PolynomialFeatures:')
    print(polynomial_features.transform(X[:, np.newaxis]))
    
    print('Coefficients of LinearRegression:')
    print(linear_regression.coef_)
    

    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, X[:, np.newaxis], y,
                             scoring="neg_mean_squared_error", cv=5)

    X_test = np.linspace(-50, 50, 10000)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model")
    plt.plot(X_test, true_fun(X_test), label="True function")
    plt.scatter(X, y, edgecolor='b', s=20, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    #plt.xlim((0, 1))
    #plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
plt.show()
```
